```python
import pandas as pd

df = pd.read_csv(path + '/data/500_Person_Gender_Height_Weight_Index.csv')
df
```

```python
# Mencari info data pada file (tipe datanya, non nul count data, nama kolom)
df.info()
```

```python
# Menghitung mean semua kolom numerik
df['Height'].mean()
```

```python
# Menghitung median semua kolom numerik
df['Height'].median()
```

```python
# Mencari modus
df['Height'].mode()
```

```python
# Menghitung Variasi & Standard Deviasi
df.var(numeric_only=True)
```

```python
# Menghitung Standar Deviasi
df.std(numeric_only=True)
```

```python
# Hitung kuartil (Q1)
q1 = df['Height'].quantile(0.25)
print("Q1 : ", q1)

# Hitung kuartil (Q3)
q3 = df['Height'].quantile(0.75)
print("Q2 : ", q3)

# Hitung IQR (Interquartil range)
iqr = q3 - q1
print("IQR : ", iqr)
```

```python
# Membuat statistika deskripsi pada type data int
df.describe()
```

```python
correlation_matrix = df.corr(numeric_only=True)

# Menampilkan hasil matriks korelasi
print("Matriks Korelasi:")
print(correlation_matrix)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Ambil data heigt
data_height = df['Height']

# Buat histogram
n, bins, patches = plt.hist(x=data_height, bins=5, color='pink', edgecolor='black')

# Tambah label
plt.title('Histogram Nilai')
plt.xlabel('Height')
plt.ylabel('Frekuensi')

# Menampilkan rentang frekuensi di sumbu x
bin_centers = 0.5 * (bins[1:] + bins[:-1])
plt.xticks(bin_centers, ['{:.0f}-{:.0f}'. format(bins[i], bins[i+1]) for i in range(len(bins)-1)])

# Menampilkan histogram nilainya
plt.show()
```
