Import library

```python
  import pandas as pd
  from sklearn.preprocessing import LabelEncoder, StandardScaler
  from sklearn.model_selection import train_test_split
  from sklearn.neighbors import KNeighborsClassifier
  from sklearn.metrics import accuracy_score
```

Encode Kolom Kategorikal

```python
df_encoded = df.copy()
label_encoders = {}

for col in df.select_dtypes(include='object').columns:
    le = LabelEncoder()
    df_encoded[col] = le.fit_transform(df[col])
    label_encoders[col] = le
```

Pisahkan Fitur & Target

```python
X = df_encoded.drop(columns=["Weather Type"])
y = df_encoded["Weather Type"]
```

Standarisasi data

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

Split dan Train data

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)
```

Mencari k terbaik

```python
best_k = 1
best_acc = 0

for k in range(1, 21):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    pred = knn.predict(X_test)
    acc = accuracy_score(y_test, pred)

    if acc > best_acc:
        best_acc = acc
        best_k = k

print("K terbaik:", best_k)
print("Akurasi:", best_acc)
```

Latih model final dengan k terbaik

```python
model = KNeighborsClassifier(n_neighbors=best_k)
model.fit(X_train, y_train)
```

Prediksi satu data baru

```python
data_baru = {
    "Temperature": 20,
    "Humidity": 70,
    "Wind Speed": 5,
    "Precipitation (%)": 40,
    "Cloud Cover": "overcast",
    "Atmospheric Pressure": 1005,
    "UV Index": 3,
    "Season": "Winter",
    "Visibility (km)": 5,
    "Location": "inland"
}

# jadikan DataFrame
df_new = pd.DataFrame([data_baru])

# encode kategori
for col in df_new.select_dtypes(include='object').columns:
    df_new[col] = label_encoders[col].transform(df_new[col])

# scale
df_new_scaled = scaler.transform(df_new)

# prediksi
pred = model.predict(df_new_scaled)
hasil = label_encoders["Weather Type"].inverse_transform(pred)

print("Prediksi Weather Type:", hasil[0])
```
