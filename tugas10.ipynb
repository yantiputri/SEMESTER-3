```python
import pandas as pd
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score
```

Membuat dataset

```python
# Data
actual = [
    "Lulus", "Lulus", "Lulus", "Lulus", "Lulus",
    "Tidak Lulus", "Tidak Lulus", "Tidak Lulus", "Tidak Lulus", "Tidak Lulus"
]

predicted = [
    "Lulus", "Lulus", "Lulus", "Lulus", "Tidak Lulus",
    "Lulus", "Tidak Lulus", "Tidak Lulus", "Tidak Lulus", "Tidak Lulus"
]
```

```python
df = pd.DataFrame({'Actual': actual, 'Predicted': predicted})
df
```

```python
# Confusion matrix
labels = ["Lulus", "Tidak Lulus"]
cm = confusion_matrix(actual, predicted, labels=labels)
print("Confusion Matrix:\n", cm)
```

```python
# Metrics
accuracy = accuracy_score(actual, predicted)
precision = precision_score(actual, predicted, pos_label="Lulus")
recall = recall_score(actual, predicted, pos_label="Lulus")

print("\nAccuracy :", accuracy)
print("Precision:", precision)
print("Recall   :", recall)
```
